```python
import seaborn as sns
import numpy as np
%matplotlib inline
tips = sns.load_dataset('tips')
# tips['day'].unique()
tips
```

```python
sns.barplot(x="sex",y="total_bill",data=tips,palette="coolwarm",estimator=np.std)
```

```python
sns.countplot(x="time", data=tips)
```

```python
sns.boxplot(x="day", y="total_bill", data=tips, hue="smoker")
```

```python
sns.violinplot(x="day", y="total_bill", data=tips)
```

```python
sns.stripplot(x="day", y="total_bill", data=tips, hue="sex", dodge=True)
```

```python
sns.violinplot(x="day", y="total_bill", data=tips, hue="sex", dodge=True)
sns.swarmplot(x="day", y="total_bill", data=tips, hue="sex", dodge=True, color="black")
```

```python
sns.catplot(x="day", y="total_bill", data=tips, kind='bar')
```
