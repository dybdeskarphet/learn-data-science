```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips = sns.load_dataset("tips")
tips
```

```python
sns.set_context(context='poster')
sns.countplot(x="sex",data=tips, )
```

Check [colormaps page of seaborn](https://seaborn.pydata.org/tutorial/color_palettes.html) for palettes.

```python
sns.lmplot(x="total_bill",y="tip", data=tips, palette="rocket", hue="sex")
```
