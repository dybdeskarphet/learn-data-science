```python
import seaborn as sns
import scipy
%matplotlib inline
tips = sns.load_dataset("tips")
flights = sns.load_dataset("flights")
tips.head()
```

```python
flights.head()
```

```python
tips.corr(numeric_only=True)
```

```python
tips_corr = tips.corr(numeric_only=True)
```

```python
sns.barplot(x="month", y="passengers", data=flights)
```

```python
sns.heatmap(tips_corr,annot=True,cmap="coolwarm")
```

```python
flight_pivot = flights.pivot_table(index="month",columns="year", values="passengers")
```

```python
sns.heatmap(flight_pivot, cmap="coolwarm"#,linecolor="black", linewidths=1
)
```
