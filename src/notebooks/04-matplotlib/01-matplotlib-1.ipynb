# Matplotlib #1

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
plt.style.use('default')
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
plt.plot(x,y)
plt.xlabel('Number')
plt.ylabel('Square of the number')
plt.title('A random plot')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
plt.tight_layout()
# axes.plot(x,y)
```

```python
axes[0].plot(x,y)
axes[1].plot(x,y)
```

# Figure Size and DPI

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='x squared')
ax.plot(x,x**3, label='x cubed')

ax.legend(loc=0)
```

# Plot Appearance

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='blue', lw=2)
# ax.set_xlim([0,1])
ax.set_ylim([0,20])
```
